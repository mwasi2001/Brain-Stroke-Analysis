```python
for i, column in enumerate(data.columns, start=1):
    plt.figure(figsize=(12, 5))

    # Custom palette with alternating colors for better visual distinction
    palette = sns.color_palette("Set2") if i % 2 == 0 else sns.color_palette("husl", len(data[column].unique()))

    # Create the count plot
    sns.countplot(data=data, x=column, palette=palette)

    # Rotate x-axis labels for readability
    plt.xticks(rotation=45, ha='right')

    # Set custom title and labels
    plt.title(f"Distribution of {column}", fontsize=14, weight='bold')
    plt.xlabel(column, fontsize=12)
    plt.ylabel("Count", fontsize=12)

    # Add count labels on top of each bar
    for p in plt.gca().patches:
        plt.gca().annotate(f'{int(p.get_height())}',
                           (p.get_x() + p.get_width() / 2., p.get_height()),
                           ha='center', va='baseline',
                           fontsize=10, color='black', weight='bold')

    # Display the plot
    plt.tight_layout()
    plt.show()
```

```python
# facts about smoking_status
never_smoked_stroke = data[(data['smoking_status'] == 'never smoked') & (data['stroke'] == 1)]['stroke'].sum()
formerly_smoked_stroke = data[(data['smoking_status'] == 'formerly smoked') & (data['stroke'] == 1)]['stroke'].sum()
smokes_stroke = data[(data['smoking_status'] == 'smokes') & (data['stroke'] == 1)]['stroke'].sum()

# show in pie chart
plt.pie([never_smoked_stroke, formerly_smoked_stroke, smokes_stroke],
        labels=['never smoked', 'formerly smoked', 'smokes'],
        autopct='%1.1f%%')
plt.title('Stroke Distribution by Smoking Status')
plt.show()
```

```python
# if 1 stroke occure if 0 not occure, show the realtion between smoking_status and stoke
stroke_counts = data.groupby('smoking_status')['stroke'].value_counts().unstack()
stroke_counts.plot(kind='bar', stacked=True)
plt.title('Stroke Distribution by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.legend(title='Stroke')
plt.show()
# Smoking is strongly linked to an increased risk of stroke
```

```python
# realtion amoung hypertension,heart_disease and stroke
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='hypertension', hue='stroke', palette='Set2')
plt.title('Hypertension vs. Stroke')
plt.xlabel('Hypertension')
plt
plt.ylabel('Count')
plt.legend(title='Stroke', loc='upper right')
plt.show()


```
